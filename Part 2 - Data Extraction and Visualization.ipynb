## Feature Extraction

```python
#Extract more ele from variety column
variety = df['Variety']
```

```python
#combine variety column into a single text
variety_text = ' '.join(variety)
```

```python
#convert all to lower case
variety_text_lower = variety_text.lower()
```

```python
#remove all punctuations
#create a list of common punctuations from string
punc_list = list(string.punctuation)
```

```python
#loop through every character in variety 
#and check if its in punc_list and join the whole thing back into a string

variety_clean = ''.join([char for char in variety_text_lower if char not in punc_list])
```

```python
#tokenise the variety string 
# This transform the text back into a list without spacing

variety_tokens = nltk.tokenize.word_tokenize(variety_clean)
variety_tokens
```

```python
#check frequency of tokens to know what to look out for
variety_counter = collections.Counter(variety_tokens)
variety_counter
```

```python
#sort the data top to bottom
variety_counter.most_common()
```

things to note:

noodle and noodles are the most recurring token

flavor, curry, chilli and spicy has two or more of the same type



```python
# create a list of spicy indicators to separate them out from variety column

is_spicy = []
hot_flavor = ['spicy',
             'curry',
             'hot',
             'tom',
             'yum',
             'chili',
             'chilli',
             'kari',
             'pedas',
             'laksa']

#usage of any to loop hot_flavor list 
#through the Variety column and execute code if found
#append 1 for spicy and 0 for non spicy to the is_spicy list
#finally create a new column to indicate if variety is spicy in df

for variety in df['Variety'].str.lower():
    if any(flavor in variety for flavor in hot_flavor):
        is_spicy.append(1)
    else:
        is_spicy.append(0)
df['isSpicy'] = is_spicy

df
```

```python
has_chicken = []

for variety in df['Variety'].str.lower():
    if ('chicken' in variety) or ('ayam' in variety):
        has_chicken.append(1)
    else:
        has_chicken.append(0)
df['hasChicken'] = has_chicken

df
```

```python
df.sample(20)
```

```python
has_beef = []

for variety in df['Variety'].str.lower():
    if ('beef' in variety) or ('cow' in variety):
        has_beef.append(1)
    else:
        has_beef.append(0)
df['hasBeef'] = has_beef

df.sample(20)
```

```python
seafood_flavor = ['seafood',
                 'fish',
                 'shrimp',
                 'prawn',
                 'crab',
                 'abalone',
                 'lobster',
                 'clams']
has_seafood = []

for variety in df['Variety'].str.lower():
    if any(flavor in variety for flavor in seafood_flavor):
        has_seafood.append(1)
    else:
        has_seafood.append(0)
        
df['hasSeafood'] = has_seafood

df.sample(20)
```

```python
# Look further into the comparison between spicy and non-spicy 

#split into two DataFrames
spicy_df = df[df['isSpicy'] == 1]
nonspicy_df = df[df['isSpicy'] == 0]
```

```python
#Check their mean

print('Spicy {}'.format(spicy_df['Stars'].mean()))
print('Non-Spicy {}'.format(nonspicy_df['Stars'].mean()))
```

```python
#Check their median

print('Spicy {}'.format(spicy_df['Stars'].median()))
print('Non-Spicy {}'.format(nonspicy_df['Stars'].median()))
```

```python
# Look at the distribution ratings(Stars) of both using a histogram
#stacking both plots
#alpha is transparency level

plt.hist(spicy_df['Stars'], alpha = 0.5, color = 'red')
plt.hist(nonspicy_df['Stars'], alpha = 0.5, color = 'black')
sns.despine()
```

This shows both to not be normally distributed ruling out a T-test for testing if it is statisticaly significant.

Althought visually they seems to behave similarly

```python
#Using seaborn to normailize

sns.histplot(spicy_df['Stars'], color = 'red')
sns.histplot(nonspicy_df['Stars'], color = 'black')
sns.despine()
```

```python
# Check if this difference is statistically significant

# Do a mann-whitney u test
# For quantities that are not normally distributed
# p-values must be less then 0.05

from scipy.stats import mannwhitneyu
```

```python
mannwhitneyu(spicy_df['Stars'],
            nonspicy_df['Stars'])
```

The Mann-Whitney U test returned a p-value of 5.317764966240481e-08 which shows that they are statistically different

so we can conclude spicy noodles tend to rate better then non-spicy
